## Linked List 

```python
class node:
    def __init__(self,data=None):
        self.data=data
        self.next=None

class linked_list:
    def __init__(self):
        self.head=node()
        self.length=0;
        self.tail = self.head

    # Adds new node containing 'data' to the end of the linked list.
    def append(self,data):
        new_node=node(data)
        cur=self.head
        while cur.next!=None:
            cur=cur.next
        cur.next=new_node
        self.length += 1
        self.tail.next = new_node
        self.tail = new_node

    # Returns the value of the node at 'index'.
    def traverseToIndex(self,index):  
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index): return cur_node.data
            curr_index += 1

    # Remove node 
    def remove(self,index):
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index-1): 
                leader = cur_node
                break;
            curr_index += 1
        
        unwantedNode = leader.next
        leader.next = unwantedNode.next
        self.length -=1

    def insertToIndex(self,index,data):
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index-1): 
                leader = cur_node
                break;
            curr_index += 1
        new_node=node(data)
        holdingIndex = leader.next;
        leader.next = new_node
        new_node.next = holdingIndex
        self.length +=1


    # Returns the length (integer) of the linked list.
    def getlength(self):
        cur=self.head
        total=0
        while cur.next!=None:
            total+=1
            cur=cur.next
        return total 

    # Prints out the linked list in traditional Python list format. 
    def display(self):
        elems=[]
        cur_node=self.head
        while cur_node.next!=None:
            cur_node=cur_node.next
            elems.append(cur_node.data)
        print(elems)

```

```python
mylst = linked_list()
print('Length of linked list-> ',mylst.length)
mylst.append(10)
mylst.append(2)
mylst.append(31)
mylst.display()
print('Length of linked list-> ',mylst.getlength())
mylst.append(23)
mylst.display()
print('Last Node Data->',mylst.tail.data)
print('First Node Data->',mylst.head.next.data)
print(mylst.traverseToIndex(2))
print('Add data to index')
mylst.insertToIndex(2,10)
mylst.display()
print('Length of linked list-> ',mylst.length)
mylst.remove(4)
mylst.display()
print('Length of linked list-> ',mylst.length)
mylst.remove(1)
mylst.display()
print('Length of linked list-> ',mylst.length)
```
